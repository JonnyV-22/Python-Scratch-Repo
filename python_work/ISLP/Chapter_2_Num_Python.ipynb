```python
# matplotlib is a common library for graphics
# The notation of plotting is not intrinsit to the language
# Using subplots() to create a figure and the axes onto which we plot our data
# matplotlib.org/stable/gallery/

# matplotlib concist of one figure and one or more axes
# Figure can be thought of the entire plotting window
# Axes contain important information about each plot such as its x and y-axis labels, title, and more

from matplotlib.pyplot import subplots
fig, ax = subplots(figsize=(8,8))

# What we did above is called unpacking the tuple in subplots into distinct variables fig and ax
# A more verbose way of doing the same above would be:
# output = subplots(figsize=(8, 8))
# fig = output[0]
# ax = output[1]


x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x,y, 'o')
fig.set_size_inches(12,4)
ax.set_xlabel("This is the x-axis")
ax.set_ylabel("This is the y-axis")
ax.set_title("X vs Y");
```

```python
# You can create a figure with several plots within it. This can be done by passing additional arguments within subplots()

figs, axes = subplots(nrows=2, ncols= 3, figsize=(15,5))
axes[0,1].plot(x,y, 'o')
axes[1,2].scatter(x, y, marker='+');
```

```python
# You can save the output of a fig by calling the savefig() method. The argument dpi is the dots per inch to determine how large the figure will be

fig.savefig("Figure.png", dpi=400)
fig.savefig("Figure.pdf", dpi=200)
```

```python
# The ax.contour() method allows to produce more sophisticated plots in order to represent 3D data similar to a topographical map
# Takes a vector of x values (first dimension), a vector of y values (second dimension), and a matrix whos elements correspond to the z value (third dimension) for each pair of (x,y)
# np.linspace returns evenly spaced numbers over a specified interval

fig, ax = subplots(figsize=(8,8))
#The below statement returns 50 numbers between negative pi and pi that are evenly spaaced
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f, levels = 45);
```

```python
# ax.imshow() is similar to ax.contour() except it produces a color-coded plot whos colors depends on the z value

fig, ax = subplots(figsize=(8,8))
ax.imshow(f);
```

```python
# Creating a 2-D array
a = np.arange(0,16).reshape((4,4))

# Since python indexing starts at 0 [1,2] refers to the second row, third column 
a[1,2]
#The above will return 6


print(a)
```

```python
# You can pass in a list specifying your suggestion as well
# This will pull back the second - fourth rows
a[[1,3]]
```

```python
# To select all rows and only the 0th and 2nd columns you can call it like this
a[:, [0,2]]
```

```python
# To get only specific columns and specific rows it gets a bit more tricky

# Trying to do a[[1,3],[0,2]] will only return [4,14]

# We can do this by creating a submatrix by subsetting the rows of A then grabbing all the rows and its columns
a[[1,3]][:,[0,2]]
```

```python
# Numpy has a more efficient way to do the above
# no.ix_() allows you to extract a submatrix using list creating a mesh object
idx = np.ix_([1,3],[0,2,3])
a[idx]
```

```python
# You can subset matrice by using slices
# 1:4:2 says include 1-4 and step by 2 so this will return only the second and fourth item 
a[1:4:2,0:3:2]
```

```python
# In numpy the boolean type is either True or False also represented as 1 and 0
keep_rows = np.zeros(a.shape[0], bool)
keep_rows
```

```python
keep_rows[[1,3]] = True
keep_rows
```

```python
# Since True and False is also represented as 1 and 0 this will return true
np.all(keep_rows == np.array([0,1,0,1]))

# The function np.all() has checked whether all entries of an array are True
```

```python
# This will return all the rtows of A
a[np.array([0,1,0,1])]
```

```python
# Contrasted to keep_rows which will only return the second and fourth rows of A where boolean eq TRUE
a[keep_rows]

# This demonstrates that Booleans and ints are treated differently by numpy
```

```python
keep_cols = np.zeros(a.shape[1], bool)
keep_cols[[0,2,3]] = True
idx_bool = np.ix_(keep_rows,keep_cols)
a[idx_bool]
```
