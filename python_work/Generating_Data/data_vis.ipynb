```python
import matplotlib.pyplot as plt

#input_values = [1,2,3,4,5,6,7]
#power_2 = [2,4,8,16,32,64,128]

x_values = range(1,5)
y_values = [2**x for x in x_values]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
#ax.plot(input_values,power_2, linewidth=3)
#ax.scatter(input_values,power_2,s=64)
ax.scatter(x_values,y_values,color='black',s=32)

'''Set chart title and lable axes'''
ax.set_title("Powers of 2", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Value of 2^X", fontsize=14)

ax.tick_params(labelsize=14)
#ax.axis([0, 15 ,0, 200_000])
#ax.ticklabel_format(style='plain')

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1,5000)
y_values = [x**3 for x in x_values]

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Greens,s=10)


'''Set chart title and lable axes'''
ax.set_title("Powers of 3", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Value of Cubed Num", fontsize=14)

ax.tick_params(labelsize=14)

plt.savefig('cubes_plot.png', bbox_inches='tight')

plt.show()
```
