```python
import random
import matplotlib.pyplot as plt

# Define the triangle vertices
pt1 = (1, 1)
pt2 = (3, 5)
pt3 = (5, 1)

# Define the number of iterations
num_iterations = 100000  # Adjust this as needed

# Initialize the starting point with a random point within the triangle
current_point_x = random.uniform(min(pt1[0], pt2[0], pt3[0]), max(pt1[0], pt2[0], pt3[0]))
current_point_y = random.uniform(min(pt1[1], pt2[1], pt3[1]), max(pt1[1], pt2[1], pt3[1]))

# Lists to store the x and y coordinates of the points
x_coords = [current_point_x]
y_coords = [current_point_y]

# Repeat the process for a specified number of iterations
for _ in range(num_iterations):
    # Choose a random vertex
    random_vertex = random.choice([pt1, pt2, pt3])

    # Calculate the halfway point between the current point and the random vertex
    current_point_x = (current_point_x + random_vertex[0]) / 2.0
    current_point_y = (current_point_y + random_vertex[1]) / 2.0

    # Append the updated point coordinates to the lists
    x_coords.append(current_point_x)
    y_coords.append(current_point_y)

plt.style.use('classic')
plt.subplots(figsize=(20,12))
# Create a scatter plot
plt.scatter(x_coords, y_coords, s=15, c='red')  # Adjust the marker size (s) as needed

# Set labels and title
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Random Points in Triangle')



# Show the plot
plt.show()
```
